## Paketzustellung [10 Punkte]

Die Paketzustellfirma wirbt damit, dass durchschnittlich 75% der Lieferungen innerhalb von 6 Tagen zugestellt werden. Sie ziehen eine Zufallsstichprobe vom Umfang 40 aus den Sendungen der letzten Wochen. Nehmen Sie an, dass die Angabe der Firma korrekt ist.

### Paketzustellung a) [5 Punkte]
Wie groß ist die Wahrscheinlichkeit, dass bei den 40 Sendungen in der Stichprobe die erste, dritte und zehnte verspätet zugestellt werden alle anderen jedoch innerhalb der 6 Tage, wenn die 6-Tagesfrist in 75% aller Fälle eingehalten wird?

Speichern Sie die Parameter der zugrundeliegenden Verteilung in den Variablen ```n, p, k``` und die Wahrscheinlichkeit in der Variablen ```wahrscheinlichkeitPaketA```.

````python
n = 40 # Stichprobengröße
p = (1 - 0.75) # Wahrscheinlichkeit dafür, dass Erfolgsfall eintritt - hier "Pakete verspätet"
k = 3 # Anzahl der Erfolge
'''
"n über k" wird hier nicht gebraucht, da nicht alle Kombination gesucht sind, sondern nur exakt eine:
Gesucht ist die Kombination, in der das erste, dritte und zehnte Paket, verspätet zugestellt wurde.

Daraus ergibt sich die eine gesuchte Kombination:
{F,T,F,T,T,T,T,T,T,F,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T,T}
 
T:=true  -> wurde binnen 6-Tagefrist zugestellt
F:=false -> wurde NICHT binnen 6-Tagefrist zugestellt

Um es verständlich darzustellen, rechne ich die '1' mit, um zu zeigen, dass es sich hier um die
Wahrscheinlichkeit für eben eine einzige Kombination handelt.
'''
wahrscheinlichkeitPaketA = 1 * (p**k) * ((1 - p)**(n-k))
print("=","{0:.8f}".format(wahrscheinlichkeitPaketA))
print("≈","{0:.5%}".format(wahrscheinlichkeitPaketA))
````

### Paketzustellung b) [5 Punkte]
Wie groß ist die Wahrscheinlichkeit, dass bei den 40 Sendungen in der Stichprobe drei Sendungen verspätet zugestellt werden alle anderen jedoch innerhalb der 6 Tage, wenn die 6-Tagesfrist in 75% aller Fälle eingehalten wird?

Speichern Sie Ihre Lösung in der Variablen ```wahrscheinlichkeitPaketB```.

````python
'''
Hier sollen wir ALLE möglichen Kombinationen berücksichtigen - also muss 'n über k' her.
Wir ersetzen ganz einfach die erste '1' aus der vorherigen Teilaufgabe durch 'n über k'.
'''
from math import comb
wahrscheinlichkeitPaketB = comb(n, k) * (p**k) * ((1-p)**(n-k))
print("=","{0:.5f}".format(wahrscheinlichkeitPaketB))
print("≈","{0:.2%}".format(wahrscheinlichkeitPaketB))

# Mit SciPy-Bibliothek -> stats.binom.pmf()
from scipy.stats import binom
P = binom.pmf(n=n, p=p, k=k)
print("\nstats.binom.pmf(n=40, k=3, p=(1-0.75)):","{0:.5f}".format(P))
print("≈","{0:.2%}".format(P))
````

## Basketballer a-e) [10 Punkte]

Eine Studie zur Verletzungsanfälligkeit von Basketballern in den zwei Zeiträumen April und September ergibt folgendes Ergebnis:

Ereignis 1 mit Wahrscheinlichkeit 70%: Der Basketballer fällt im April **nicht** aus.

Ereignis 2 mit Wahrscheinlichkeit 10%: Der Basketballer fällt im September **nicht** aus, wenn wir wissen, dass der Basketballer im April bereits ausgefallen ist.

Ereignis 3 mit Wahrscheinlichkeit 63%: Der Basketballer ist im September und im April gesund.

Verwenden Sie dazu die Konvention:

- $P(A)$ bedeutet der Basketballer fällt im April aus
- $P(A^C)$ bedeutet der Basketballer fällt im April **nicht** aus (ist gesund)
- $P(S)$ bedeutet der Basketballer fällt im September aus
- $P(S^C)$ bedeutet der Basketballer fällt im September **nicht** aus (ist gesund)


*Hinweis:*  Ein Baum auf dem Schmierblatt ist hilfreich

*Geben Sie Ihre Lösungen mit einer Genauigkeit von 4 Stellen nach dem Komma an.*

### Basketballer a) [2 Punkte]

Wie hoch ist die Wahrscheinlichkeit dafür, dass der Basketballer im September ausfällt?

Speichern Sie Ihre Lösung in der Variablen ```WahrscheinlichkeitA```.

````python
'''
Gesucht ist hier die totale Wahrscheinlichkeit für S -> P(S)

    P(S) = P(S|A)*P(A) + P(S|Ā)*P(Ā)

Begriffe zur totalen Wahrscheinlichkeit: paarweise disjunkte Zerlegung bzw. Partition
'''
PvonSgegebenA = 0.9
PvonA = 0.3
PvonSgegebenNichtA = 0.37
PvonNichtA = 0.7

# round() benutzen, um die geforderte Genauigkeit von 4 Stellen nach dem Komma zu erhalten
WahrscheinlichkeitA = round(PvonSgegebenA*PvonA + PvonSgegebenNichtA*PvonNichtA, 4)

print(WahrscheinlichkeitA,"= {0:.2%}".format(WahrscheinlichkeitA))
````

Speichern Sie Ihre Lösung in der Variablen ```WahrscheinlichkeitB```.

````python
'''
Hier brauchen wir den Satz von Bayes und die totale Wahrscheinlichkeit "P(S)"
aus der vorherigen Teilaufgabe.

Satz von Bayes:
    
    P(A|S) = P(S|A)*P(A) / P(S)

Die totale Wahrscheinlichkeit P(S) haben wir bereits in der a) ausgerechnet.
'''
PvonSgegebenA = 0.9
PvonA = 0.3
PvonS = WahrscheinlichkeitA

# Auch hier wieder 'round()', aus gegebenem Anlass.
WahrscheinlichkeitB = round((PvonSgegebenA * PvonA) / PvonS, 4)

print(WahrscheinlichkeitB,"= {0:.2%}".format(WahrscheinlichkeitB))
````
